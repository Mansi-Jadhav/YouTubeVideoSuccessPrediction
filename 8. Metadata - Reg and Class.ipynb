## Import and set seeds

```python
import random
import torch
from pathlib import Path
import joblib

# Set seeds for reproducibility
def set_seed(seed=42):
    # Python & NumPy
    random.seed(seed)
    np.random.seed(seed)
    
    # PyTorch
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)  # For multi-GPU
    
    # cuDNN (may impact performance)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
    
    # Environment (for some libraries)
    os.environ['PYTHONHASHSEED'] = str(seed)

MODEL_DIR = Path("models_final")
MODEL_DIR.mkdir(parents=True, exist_ok=True)
RESULTS_DIR = Path("results_final")
RESULTS_DIR.mkdir(parents=True, exist_ok=True)
```

# Data Pre-processing

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, FunctionTransformer
from sklearn.base import clone
import numpy as np

cat_cols = ['duration_type','channel_size','country','video_category_id']   # keep this ORDER fixed
num_cols = ['subscriber_count','prev_video_count','avg_views_prev10','avg_likes_prev10','avg_comments_prev10','upload_hour']
feature_cols = ['country','video_category_id','duration_type','subscriber_count','upload_hour',
                'channel_size','prev_video_count','avg_views_prev10','avg_likes_prev10','avg_comments_prev10']

def make_preprocessor():
    num = Pipeline([
        ("log1p", FunctionTransformer(np.log1p, feature_names_out="one-to-one")),  # Transform the numerical columns to log1p
        ("scale", StandardScaler()),     # Standardize the transformed columns
    ])
    return ColumnTransformer([
        ("onehot", OneHotEncoder(handle_unknown="ignore"), cat_cols),  # One hot encode the categorical columns
        ("num", num, num_cols),
    ])

def make_pipeline(estimator):
    return Pipeline([("prep", make_preprocessor()), ("model", estimator)])  # Pipeline of preprocessor and model
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from scipy.stats import pearsonr
import numpy as np
import os
import joblib

def metrics(y_true, y_pred, label=""):
    mse  = mean_squared_error(y_true, y_pred) # Return Mean Squared Error
    rmse = np.sqrt(mse)   # Root Mean Squared Error
    mae  = mean_absolute_error(y_true, y_pred)  # Mean Absolute Error
    r2   = r2_score(y_true, y_pred)    # R2 score
    pc   = pearsonr(y_true, y_pred)[0]   # Pearson R
    print(f"{label:>10} | MSE {mse:8.4f}  RMSE {rmse:6.3f} "
          f" MAE {mae:6.3f}  R¬≤ {r2:6.3f}  Pearson r {pc:6.3f}")
    return mse, rmse, r2, pc
set_seed(42)

df = pd.read_csv('FinalDataset.csv')

train_video_ids = set(os.listdir("Person-Independent_Split\\train"))
train_df = df[df["video_id"].isin(train_video_ids)].reset_index(drop=True)

val_video_ids = set(os.listdir("Person-Independent_Split\\val"))
val_df = df[df["video_id"].isin(val_video_ids)].reset_index(drop=True)

target_col = 'log_view_count'

X_train = train_df[feature_cols]
y_train = train_df[target_col]

X_val = val_df[feature_cols]
y_val = val_df[target_col]

pipe = make_pipeline(SVR(kernel="linear"))   # Preprocessor + Model 
pipe.fit(X_train, y_train)  # Fit on the train data
y_train_pred = pipe.predict(X_train)   # To calculate MSE on train
y_val_pred   = pipe.predict(X_val)    # Predict on the validation data

print("Train Metrics:")
train_mse, train_rmse, train_r2, train_pc = metrics(y_train, y_train_pred)
print("Validation Metrics:")
val_mse, val_rmse, val_r2, val_pc = metrics(y_val, y_val_pred)

out_path = MODEL_DIR / "metadata_reg_ind.pkl"
joblib.dump(pipe, out_path)        # Save the pipeline -> Can be used for test
print("\nModel saved ‚Üí metadata_reg_ind.pkl")
pd.DataFrame({"video_id": val_df.video_id, "y_true": y_val, "y_pred": y_val_pred}).to_csv(
    RESULTS_DIR / "metadata_reg_val_ind.csv", index=False)                  # Save predictions for late fusion
print("\nPredictions saved for late-fusion -> metadata_reg_ind_val.csv")
```

# Test Predictions

## Regression

```python
import os, joblib, numpy as np, pandas as pd
from pathlib import Path

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler, FunctionTransformer
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from scipy.stats import pearsonr

CSV_PATH   = "FinalDataset.csv"
MODEL_DIR  = Path("models_final") 
MODEL_DIR.mkdir(exist_ok=True)

# saved best-after-val models
MODEL_PATHS = {
    "ind": MODEL_DIR / "metadata_reg_ind.pkl",
    "dep": MODEL_DIR / "metadata_reg_dep.pkl",
}

# directories containing ids for each split
SPLIT_DIRS = {
    "ind": {
        "base": "Person-Independent_Split",
        "train": "train", "val": "val", "test": "test",
    },
    "dep": {
        "base": "Person-Dependent_Split",
        "train": "train", "val": "val", "test": "test",
    }
}

# features / target
FEATURE_COLS = [
    "country", "video_category_id", "duration_type", "subscriber_count", "upload_hour",
    "channel_size", "prev_video_count", "avg_views_prev10", "avg_likes_prev10", "avg_comments_prev10"
]
TARGET = "log_view_count"

CAT_COLS = ["country", "video_category_id", "duration_type", "channel_size"]
NUM_COLS = ["subscriber_count", "prev_video_count", "avg_views_prev10",
            "avg_likes_prev10", "avg_comments_prev10", "upload_hour"]

# ----------------------- Utils ------------------------
def read_ids_from_dir(dir_path: str):
    # files are named like "<video_id>.*" ‚Äî strip extensions
    return set(os.path.splitext(fn)[0] for fn in os.listdir(dir_path))

def make_preprocessor():
    numeric_pipe = Pipeline([
        ("log1p",  FunctionTransformer(np.log1p, feature_names_out="one-to-one")),
        ("scale",  StandardScaler()),
    ])
    return ColumnTransformer([
        ("cat", OneHotEncoder(handle_unknown="ignore"), CAT_COLS),
        ("num", numeric_pipe, NUM_COLS),
    ])

def metrics_report(y_true, y_pred, tag="TEST"):
    mse  = mean_squared_error(y_true, y_pred)
    rmse = float(np.sqrt(mse))
    mae  = mean_absolute_error(y_true, y_pred)
    r2   = r2_score(y_true, y_pred)
    pr   = pearsonr(y_true, y_pred)[0]
    print(f"{tag}: MSE {mse:.4f} | RMSE {rmse:.3f} | MAE {mae:.3f} | R¬≤ {r2:.3f} | Pearson r {pr:.3f}")
    return dict(mse=mse, rmse=rmse, mae=mae, r2=r2, pearson=pr)

def expm1_safe(arr): return np.expm1(np.asarray(arr, dtype=float))

# ----------------------- Data for a split ------------------------
def get_trainval_test_df(split_tag: str):
    cfg = SPLIT_DIRS[split_tag]
    base = cfg["base"]
    train_ids = read_ids_from_dir(os.path.join(base, cfg["train"]))
    val_ids   = read_ids_from_dir(os.path.join(base, cfg["val"]))
    test_ids  = read_ids_from_dir(os.path.join(base, cfg["test"]))

    df = pd.read_csv(CSV_PATH)
    trainval_ids = train_ids | val_ids

    trainval_df = df[df.video_id.isin(trainval_ids)].reset_index(drop=True)
    test_df     = df[df.video_id.isin(test_ids)].reset_index(drop=True)

    # keep only needed columns for safety
    cols_need = ["video_id", TARGET] + FEATURE_COLS
    trainval_df = trainval_df[cols_need].dropna().reset_index(drop=True)
    test_df     = test_df[cols_need].dropna().reset_index(drop=True)
    return trainval_df, test_df

# ----------------------- Runner ------------------------
def retrain_on_trainval_then_test(split_tag: str):
    print(f"\n===== [{split_tag.upper()}] Metadata SVR: train+val -> test =====")
    model_path = MODEL_PATHS[split_tag]
    assert model_path.exists(), f"Missing model: {model_path}"

    # load data
    trainval_df, test_df = get_trainval_test_df(split_tag)
    print(f"Rows ‚Äî train+val: {len(trainval_df)} | test: {len(test_df)}")

    X_trv, y_trv = trainval_df[FEATURE_COLS], trainval_df[TARGET]
    X_tst, y_tst = test_df[FEATURE_COLS],     test_df[TARGET]

    # load saved model (pipeline or bare estimator)
    saved = joblib.load(model_path)

    if hasattr(saved, "named_steps"):  # Pipeline: just refit on train+val
        pipe = saved
        pipe.fit(X_trv, y_trv)
        y_pred = pipe.predict(X_tst)

    else:
        # Bare estimator: rebuild preprocessor, fit on train+val, wrap
        preproc = make_preprocessor()
        X_trv_proc = preproc.fit_transform(X_trv)
        X_tst_proc = preproc.transform(X_tst)
        estimator = saved
        estimator.fit(X_trv_proc, y_trv)
        # wrap for saving consistency
        pipe = Pipeline([("prep", preproc), ("svr", estimator)])
        y_pred = estimator.predict(X_tst_proc)

    # ----- metrics (log space) -----
    _ = metrics_report(y_tst, y_pred, tag="TEST (log)")
    # also report view-space MSE for intuition
    mse_views = float(np.mean((expm1_safe(y_tst) - expm1_safe(y_pred))**2))
    print(f"TEST (views): MSE {mse_views:.2f}")

    # ----- save predictions -----
    preds_csv = RESULTS_DIR / f"metadata_reg_test_{split_tag}.csv"
    out = test_df[["video_id"]].copy()
    out["y_true"]   = y_tst
    out["y_pred"]   = y_pred
    out["pred_views"] = expm1_safe(y_pred)
    out.to_csv(preds_csv, index=False)
    print(f"Saved test predictions ‚Üí {preds_csv}")

    # ----- save refit model (train+val) -----
    refit_path = MODEL_DIR / f"metadata_reg_final_{split_tag}.pkl"
    joblib.dump(pipe, refit_path)
    print(f"Saved refit model (train+val) ‚Üí {refit_path}")

    return {
        "preds_csv": preds_csv,
        "refit_model": refit_path,
    }

# ----------------------- Run both splits ------------------------
res_ind = retrain_on_trainval_then_test("ind")
res_dep = retrain_on_trainval_then_test("dep")
```

## Classification

```python
# ======================= Metadata SVC: Train+Val -> Test (robust) =======================
import os, joblib, numpy as np, pandas as pd
from pathlib import Path

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler, FunctionTransformer
from sklearn.metrics import accuracy_score, f1_score, precision_recall_fscore_support, classification_report, confusion_matrix

# ----------------------- Config -----------------------
CSV_PATH   = "FinalDataset.csv"
MODEL_DIR  = Path("models") 
MODEL_DIR.mkdir(exist_ok=True)

# saved best-after-val classifiers
MODEL_PATHS = {
    "ind": MODEL_DIR / "metadata_cls_ind.pkl",
    "dep": MODEL_DIR / "metadata_cls_dep.pkl",
}

# split folders
SPLIT_DIRS = {
    "ind": {"base": "Person-Independent_Split", "train": "train", "val": "val", "test": "test"},
    "dep": {"base": "Person-Dependent_Split",   "train": "train", "val": "val", "test": "test"},
}

# label columns
LABEL_COL_MAP = {"ind": "view_range_enc_ind", "dep": "view_range_enc_dep"}

# features
FEATURE_COLS = [
    "country", "video_category_id", "duration_type", "subscriber_count", "upload_hour",
    "channel_size", "prev_video_count", "avg_views_prev10", "avg_likes_prev10", "avg_comments_prev10"
]
CAT_COLS = ["country", "video_category_id", "duration_type", "channel_size"]
NUM_COLS = ["subscriber_count", "prev_video_count", "avg_views_prev10",
            "avg_likes_prev10", "avg_comments_prev10", "upload_hour"]

# ----------------------- Utils ------------------------
def read_ids_from_dir(dir_path: str):
    return set(os.path.splitext(fn)[0] for fn in os.listdir(dir_path))

def make_preprocessor():
    numeric_pipe = Pipeline([
        ("log1p",  FunctionTransformer(np.log1p, feature_names_out="one-to-one")),
        ("scale",  StandardScaler()),
    ])
    return ColumnTransformer([
        ("cat", OneHotEncoder(handle_unknown="ignore"), CAT_COLS),
        ("num", numeric_pipe, NUM_COLS),
    ])

def get_trainval_test_df(split_tag: str):
    cfg = SPLIT_DIRS[split_tag]
    base = cfg["base"]
    train_ids = read_ids_from_dir(os.path.join(base, cfg["train"]))
    val_ids   = read_ids_from_dir(os.path.join(base, cfg["val"]))
    test_ids  = read_ids_from_dir(os.path.join(base, cfg["test"]))

    df = pd.read_csv(CSV_PATH)
    label_col = LABEL_COL_MAP[split_tag]

    trainval_ids = train_ids | val_ids
    cols_need = ["video_id", label_col] + FEATURE_COLS

    trainval_df = df[df.video_id.isin(trainval_ids)][cols_need].dropna().reset_index(drop=True)
    test_df     = df[df.video_id.isin(test_ids)][cols_need].dropna().reset_index(drop=True)

    # ensure integer labels
    trainval_df[label_col] = trainval_df[label_col].astype(int)
    test_df[label_col]     = test_df[label_col].astype(int)
    return trainval_df, test_df, label_col

def model_supports_proba(est):
    if hasattr(est, "predict_proba"): 
        return True
    if hasattr(est, "named_steps"):
        return hasattr(est.steps[-1][1], "predict_proba")
    return False

def last_estimator(est):
    """Return the final estimator whether `est` is a Pipeline or a bare estimator."""
    return est.steps[-1][1] if hasattr(est, "steps") else est

# ----------------------- Runner ------------------------
def retrain_on_trainval_then_test_cls(split_tag: str):
    print(f"\n===== [{split_tag.upper()}] Metadata SVC: train+val -> test =====")
    model_path = MODEL_PATHS[split_tag]
    assert model_path.exists(), f"Missing model: {model_path}"

    trainval_df, test_df, label_col = get_trainval_test_df(split_tag)
    print(f"Rows ‚Äî train+val: {len(trainval_df)} | test: {len(test_df)}")

    X_trv, y_trv = trainval_df[FEATURE_COLS], trainval_df[label_col]
    X_tst, y_tst = test_df[FEATURE_COLS],     test_df[label_col]

    saved = joblib.load(model_path)

    # Refit on train‚à™val
    if hasattr(saved, "named_steps"):           # Pipeline
        pipe = saved
        pipe.fit(X_trv, y_trv)
        y_pred = pipe.predict(X_tst)
        # probs or decision scores via pipeline (so preprocessing is applied)
        y_proba, scores = None, None
        if model_supports_proba(pipe):
            y_proba = pipe.predict_proba(X_tst)
        else:
            scores = pipe.decision_function(X_tst)
    else:
        # Bare estimator -> rebuild preproc on train+val and wrap
        preproc = make_preprocessor()
        X_trv_proc = preproc.fit_transform(X_trv)
        X_tst_proc = preproc.transform(X_tst)
        clf = saved
        clf.fit(X_trv_proc, y_trv)
        y_pred = clf.predict(X_tst_proc)
        y_proba, scores = (clf.predict_proba(X_tst_proc), None) if model_supports_proba(clf) else (None, clf.decision_function(X_tst_proc))
        pipe = Pipeline([("prep", preproc), ("clf", clf)])  # wrapped for saving

    # ----- metrics -----
    acc  = accuracy_score(y_tst, y_pred)
    f1m  = f1_score(y_tst, y_pred, average="macro")
    prec, rec, f1_per, _ = precision_recall_fscore_support(y_tst, y_pred, average=None, zero_division=0)
    print(f"TEST: Acc {acc:.4f} | Macro-F1 {f1m:.4f}")
    for cls, (p, r, f) in enumerate(zip(prec, rec, f1_per)):
        print(f"  class {cls}: P={p:.3f} R={r:.3f} F1={f:.3f}")
    print("\nClassification report:\n", classification_report(y_tst, y_pred, digits=3, zero_division=0))
    print("Confusion matrix:\n", confusion_matrix(y_tst, y_pred))

    # ----- assemble and save predictions -----
    preds_csv = RESULTS_DIR / f"metadata_cls_test_{split_tag}.csv"

    out = pd.DataFrame({
        "video_id": test_df["video_id"].values,
        "y_true": y_tst.values,
        "y_pred": y_pred
    })

    # classes from the last estimator (no hardcoded 'svc' key)
    classes = last_estimator(pipe).classes_

    if y_proba is not None:
        for i, c in enumerate(classes):
            out[f"prob_c{int(c)}"] = y_proba[:, i]
    else:
        scores = np.asarray(scores)
        if scores.ndim == 1:  # binary edge-case
            scores = np.c_[-scores, scores]
        for i, c in enumerate(classes[:scores.shape[1]]):
            out[f"score_class_{int(c)}"] = scores[:, i]

    out.to_csv(preds_csv, index=False)
    print(f"Saved test predictions ‚Üí {preds_csv}")

    # ----- save refit model (train+val) -----
    refit_path = MODEL_DIR / f"metadata_cls_final_{split_tag}.pkl"
    joblib.dump(pipe, refit_path)
    print(f"Saved refit model (train+val) ‚Üí {refit_path}")

    return {"preds_csv": preds_csv, "refit_model": refit_path, "acc": acc, "macro_f1": f1m}

# ----------------------- Run both splits ------------------------
res_ind = retrain_on_trainval_then_test_cls("ind")
res_dep = retrain_on_trainval_then_test_cls("dep")
```
